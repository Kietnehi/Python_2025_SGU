```python
# =========================
# BÀI TẬP 1 - GIẢI BẰNG PYTHON
# =========================

from math import isclose, sqrt
from typing import List, Tuple


# --------- TIỆN ÍCH CHUNG ---------
def is_prime(n: int) -> bool:
    """Kiểm tra số nguyên tố."""
    if n < 2:
        return False
    if n % 2 == 0:
        return n == 2
    r = int(sqrt(n))
    for p in range(3, r + 1, 2):
        if n % p == 0:
            return False
    return True


# --------- 1) Bài toán số học: 2 số (nguyên/thực) ---------
def arithmetic_two_numbers(a: float, b: float) -> dict:
    """Trả về tổng/hiệu/tích/thương (nếu b != 0)."""
    out = {
        "tong": a + b,
        "hieu": a - b,
        "tich": a * b,
        "thuong": None if isclose(b, 0.0) else a / b,
    }
    return out


# --------- 2) Bài về số nguyên với N ---------
def check_prime_N(N: int) -> bool:
    """a) N có phải số nguyên tố?"""
    return is_prime(N)


def list_1_to_N_and_sum(N: int) -> Tuple[List[int], int]:
    """b) Xuất các số 1..N và tổng của chúng."""
    arr = list(range(1, N + 1))
    return arr, sum(arr)


def primes_to_N(N: int) -> List[int]:
    """c) Các số nguyên tố từ 1..N (bao gồm N nếu là NT)."""
    return [x for x in range(2, N + 1) if is_prime(x)]


def first_N_primes(N: int) -> List[int]:
    """d) N số nguyên tố đầu tiên."""
    res, x = [], 2
    while len(res) < N:
        if is_prime(x):
            res.append(x)
        x += 1
    return res


# --------- 3) Bài về mảng (list số nguyên) ---------
def array_elements_and_sum(a: List[int]) -> Tuple[List[int], int]:
    """a) Liệt kê phần tử và tổng."""
    return a[:], sum(a)


def array_prime_elements(a: List[int]) -> List[int]:
    """b) Các phần tử là số nguyên tố trong mảng a."""
    return [x for x in a if is_prime(x)]


def array_append(a: List[int], value: int) -> List[int]:
    """c) Thêm 1 phần tử vào mảng, trả bản sao mới."""
    b = a[:]
    b.append(value)
    return b


def array_delete_k(a: List[int], k: int) -> List[int]:
    """
    d) Xoá phần tử thứ k của mảng a (chỉ số 0-based).
       Nếu k ngoài phạm vi -> trả về bản sao không đổi.
    """
    b = a[:]
    if 0 <= k < len(b):
        del b[k]
    return b


def array_find_x_positions(a: List[int], x: int) -> List[int]:
    """
    e) Tìm mọi vị trí (0-based) của x trong mảng a.
       Nếu không có, trả list rỗng.
    """
    return [i for i, v in enumerate(a) if v == x]


# --------- 4) Bài về chuỗi ---------
def strlen_s(s: str) -> int:
    """a) Độ dài chuỗi s (số ký tự)."""
    return len(s)


def strip_extra_spaces(s: str) -> str:
    """
    b) Xoá khoảng trắng thừa (cắt đầu/cuối, và giữa các từ chỉ còn 1 space).
    """
    return " ".join(s.split())


def word_count_and_print_lines(s: str) -> Tuple[int, List[str]]:
    """
    c) Đếm số từ của s và trả danh sách mỗi từ 1 dòng.
       Quy ước tách từ bằng split() (theo khoảng trắng).
    """
    words = s.split()
    return len(words), words


def left_right_k_chars(s: str, k: int) -> Tuple[str, str]:
    """
    d) Trả về (k ký tự bên trái, k ký tự bên phải) của s.
       Nếu k > len(s) thì cắt vừa đủ.
    """
    k = max(0, k)
    left = s[:k]
    right = s[-k:] if k > 0 else ""
    return left, right


def substring_from_k_with_len_n(s: str, k: int, n: int) -> str:
    """
    e) Xuất n ký tự của s kể từ vị trí k (0-based).
       Nếu k ngoài phạm vi -> trả chuỗi rỗng.
    """
    if k < 0 or k >= len(s) or n <= 0:
        return ""
    return s[k : k + n]


# =========================
# DEMO NHANH (có thể sửa/xoá tuỳ ý)
# =========================
if __name__ == "__main__":
    # 1) Số học
    print("1) Số học 2 số:", arithmetic_two_numbers(12, 5))

    # 2) N nguyên
    N = 20
    print("2a) N có phải NT?", N, check_prime_N(N))
    arr_1N, s_1N = list_1_to_N_and_sum(N)
    print("2b) 1..N:", arr_1N, "Tổng:", s_1N)
    print("2c) NT 1..N:", primes_to_N(N))
    print("2d) 10 NT đầu tiên:", first_N_primes(10))

    # 3) Mảng
    a = [2, 3, 4, 5, 5, 10, 11]
    print("3a) Mảng & tổng:", array_elements_and_sum(a))
    print("3b) Phần tử NT:", array_prime_elements(a))
    print("3c) Thêm 7:", array_append(a, 7))
    print("3d) Xoá phần tử thứ k=2:", array_delete_k(a, 2))
    print("3e) Vị trí x=5:", array_find_x_positions(a, 5))

    # 4) Chuỗi
    s = "   Python   là   vui   quá   "
    print("4a) Độ dài:", strlen_s(s))
    print("4b) Strip extra:", strip_extra_spaces(s))
    cnt, words = word_count_and_print_lines(s)
    print("4c) Số từ:", cnt, "| Mỗi từ 1 dòng:", words)
    print("4d) Trái/phải k=4:", left_right_k_chars("abcdefg", 4))
    print("4e) Sub từ k=2, n=5:", substring_from_k_with_len_n("0123456789", 2, 5))
```

```python
# qlsv_simple.py
# Bản đơn giản, dễ gõ – vẫn có kế thừa, đa hình, "trừu tượng" (NotImplementedError), đóng gói.

import json

# ========================
# LỚP CƠ SỞ (giả "abstract")
# ========================
class StudentBase:
    def __init__(self, sid, name, year):
        self._id = sid.strip()      # "đóng gói" kiểu nhẹ (protected)
        self._name = name.strip()
        self._year = int(year)
        self._courses = []          # mỗi phần tử: {"name": str, "credits": int, "letter": "A..F"}

    # ---- hành vi chung ----
    def add_course(self, name, credits, letter):
        name = name.strip()
        letter = letter.strip().upper()
        if letter not in ("A", "B", "C", "D", "F"):
            raise ValueError("Điểm phải là A/B/C/D/F")
        if int(credits) <= 0:
            raise ValueError("Tín chỉ > 0")
        self._courses.append({"name": name, "credits": int(credits), "letter": letter})

    def gpa(self):
        if not self._courses:
            return 0.0
        pts = {"A":4.0, "B":3.0, "C":2.0, "D":1.0, "F":0.0}
        total_p = 0.0
        total_c = 0
        for c in self._courses:
            total_p += pts[c["letter"]] * c["credits"]
            total_c += c["credits"]
        return round(total_p / total_c, 2) if total_c else 0.0

    # ---- đa hình / "trừu tượng" đơn giản ----
    def student_type(self):
        raise NotImplementedError

    def tuition(self):
        raise NotImplementedError

    def status(self):
        raise NotImplementedError

    # ---- tiện ích xuất/nhập (đơn giản) ----
    def to_dict(self):
        return {
            "type": self.student_type(),
            "id": self._id,
            "name": self._name,
            "year": self._year,
            "courses": list(self._courses),
        }

    @staticmethod
    def from_dict(d):
        t = (d.get("type") or "").lower()
        if t == "undergraduate":
            s = Undergraduate(d["id"], d["name"], d["year"])
        elif t == "graduate":
            s = Graduate(d["id"], d["name"], d["year"], d.get("advisor", ""))
        else:
            raise ValueError("Loại SV không hợp lệ")
        for c in d.get("courses", []):
            s.add_course(c["name"], c["credits"], c["letter"])
        return s


# ========================
# LỚP CON: ĐẠI HỌC
# ========================
class Undergraduate(StudentBase):
    BASE_PER_CREDIT = 350_000
    SENIOR_FACTOR = 1.10  # năm 4+ phụ 10%

    def student_type(self):
        return "Undergraduate"

    def tuition(self):
        credits = sum(c["credits"] for c in self._courses)
        fee = credits * self.BASE_PER_CREDIT
        if self._year >= 4:
            fee *= self.SENIOR_FACTOR
        return round(fee, 2)

    def status(self):
        g = self.gpa()
        if g >= 3.2: return "Honors (ĐH)"
        if g >= 2.0: return "Good (ĐH)"
        return "Probation (ĐH)"


# ========================
# LỚP CON: CAO HỌC
# ========================
class Graduate(StudentBase):
    BASE_PER_CREDIT = 600_000
    MERIT_DISCOUNT = 0.95  # GPA >= 3.5 giảm 5%

    def __init__(self, sid, name, year, advisor=""):
        super().__init__(sid, name, year)
        self._advisor = advisor.strip()

    def student_type(self):
        return "Graduate"

    def tuition(self):
        credits = sum(c["credits"] for c in self._courses)
        fee = credits * self.BASE_PER_CREDIT
        if self.gpa() >= 3.5:
            fee *= self.MERIT_DISCOUNT
        return round(fee, 2)

    def status(self):
        g = self.gpa()
        if g >= 3.5: return "Distinction (CH)"
        if g >= 2.5: return "Good (CH)"
        return "Probation (CH)"

    def to_dict(self):
        d = super().to_dict()
        d["advisor"] = self._advisor
        return d


# ========================
# QUẢN LÝ DANH SÁCH
# ========================
class StudentManager:
    def __init__(self):
        self._list = []

    # CRUD
    def add(self, s):
        if self.find_by_id(s._id):
            raise ValueError("ID đã tồn tại")
        self._list.append(s)

    def remove(self, sid):
        s = self.find_by_id(sid)
        if s:
            self._list.remove(s)
            return True
        return False

    def find_by_id(self, sid):
        sid = sid.strip()
        for s in self._list:
            if s._id == sid:
                return s
        return None

    def search_by_name(self, kw):
        kw = kw.strip().lower()
        return [s for s in self._list if kw in s._name.lower()]

    # Sắp xếp
    def sort_by_gpa(self, desc=True):
        return sorted(self._list, key=lambda x: x.gpa(), reverse=desc)

    def sort_by_name(self):
        return sorted(self._list, key=lambda x: x._name.lower())

    # Báo cáo nhanh
    def report_rows(self, data=None):
        data = data if data is not None else self._list
        rows = []
        for s in data:
            rows.append({
                "id": s._id,
                "name": s._name,
                "type": s.student_type(),
                "year": s._year,
                "gpa": s.gpa(),
                "tuition": s.tuition(),
                "status": s.status(),
                "num_courses": len(s._courses),
            })
        return rows

    # In bảng ngắn gọn
    def print_table(self, rows):
        if not rows:
            print("Không có dữ liệu.")
            return
        keys = ["id","name","type","year","gpa","tuition","status","num_courses"]
        widths = {k: max(len(k), max(len(str(r[k])) for r in rows)) for k in keys}
        def line(ch="-"):
            print("+", end="")
            for k in keys:
                print(ch*(widths[k]+2) + "+", end="")
            print()
        line("=")
        print("| " + " | ".join(k.upper().ljust(widths[k]) for k in keys) + " |")
        line("=")
        for r in rows:
            print("| " + " | ".join(str(r[k]).ljust(widths[k]) for k in keys) + " |")
        line("=")

    # Lưu/đọc JSON (rất gọn)
    def save(self, path):
        data = {"students":[s.to_dict() for s in self._list]}
        with open(path, "w", encoding="utf-8") as f:
            json.dump(data, f, ensure_ascii=False, indent=2)

    def load(self, path):
        with open(path, "r", encoding="utf-8") as f:
            data = json.load(f)
        self._list = [StudentBase.from_dict(d) for d in data.get("students", [])]


# ========================
# MENU CỰC GỌN
# ========================
def input_int(msg, lo=None, hi=None):
    while True:
        try:
            v = int(input(msg))
            if lo is not None and v < lo: print(f">= {lo}"); continue
            if hi is not None and v > hi: print(f"<= {hi}"); continue
            return v
        except:
            print("Nhập số hợp lệ!")

def create_student():
    sid = input("ID: ").strip()
    name = input("Tên: ").strip()
    year = input_int("Năm (1..6): ", 1, 6)
    t = input("Loại (u=ĐH, g=CH): ").strip().lower()
    if t == "u":
        s = Undergraduate(sid, name, year)
    elif t == "g":
        adv = input("Người hướng dẫn (bỏ trống nếu không): ").strip()
        s = Graduate(sid, name, year, adv)
    else:
        raise ValueError("Loại không hợp lệ")
    # thêm môn tuỳ chọn
    while True:
        more = input("Thêm môn? (y/n): ").strip().lower()
        if more != "y": break
        cname = input("  Tên môn: ").strip()
        ccre = input_int("  Tín chỉ: ", 1, None)
        clet = input("  Điểm (A/B/C/D/F): ").strip().upper()
        s.add_course(cname, ccre, clet)
    return s

def show_detail(s):
    print("\n--- CHI TIẾT ---")
    print("ID      :", s._id)
    print("Tên     :", s._name)
    print("Loại    :", s.student_type())
    if isinstance(s, Graduate):
        print("Advisor :", s._advisor)
    print("Năm     :", s._year)
    print("GPA     :", s.gpa())
    print("Học phí :", s.tuition(), "VND")
    print("Trạng thái:", s.status())
    print("Môn học :", len(s._courses))
    for i, c in enumerate(s._courses, 1):
        print(f"  {i:02d}. {c['name']} - {c['credits']} tín chỉ - {c['letter']}")

def main():
    mgr = StudentManager()

    # demo sẵn
    u = Undergraduate("SV001", "Nguyễn Văn A", 3)
    u.add_course("CTDL", 4, "B")
    u.add_course("Toán Rời Rạc", 3, "A")
    mgr.add(u)

    g = Graduate("SV100", "Trần Thị B", 1, "GS. Hồ")
    g.add_course("ML Nâng Cao", 3, "A")
    g.add_course("NLP", 3, "A")
    mgr.add(g)

    MENU = """
===== MENU QLSV (đơn giản) =====
1) Thêm sinh viên
2) Xoá sinh viên
3) Thêm môn cho SV
4) Tìm theo ID (in chi tiết)
5) Tìm theo tên (bảng)
6) Liệt kê theo GPA
7) Liệt kê theo TÊN
8) Lưu JSON
9) Tải JSON
0) Thoát
Chọn: """
    while True:
        print(MENU) 
        ch = input(MENU).strip()
        try:
            if ch == "1":
                s = create_student()
                mgr.add(s)
                print("OK, đã thêm.")
            elif ch == "2":
                sid = input("ID: ").strip()
                print("Đã xoá." if mgr.remove(sid) else "Không tìm thấy.")
            elif ch == "3":
                sid = input("ID: ").strip()
                s = mgr.find_by_id(sid)
                if not s: print("Không thấy."); continue
                cname = input("  Tên môn: ").strip()
                ccre = input_int("  Tín chỉ: ", 1, None)
                clet = input("  Điểm (A/B/C/D/F): ").strip().upper()
                s.add_course(cname, ccre, clet)
                print("Đã thêm môn.")
            elif ch == "4":
                sid = input("ID: ").strip()
                s = mgr.find_by_id(sid)
                show_detail(s) if s else print("Không thấy.")
            elif ch == "5":
                kw = input("Từ khoá tên: ").strip()
                mgr.print_table(mgr.report_rows(mgr.search_by_name(kw)))
            elif ch == "6":
                mgr.print_table(mgr.report_rows(mgr.sort_by_gpa(True)))
            elif ch == "7":
                mgr.print_table(mgr.report_rows(mgr.sort_by_name()))
            elif ch == "8":
                p = input("File (vd: students.json): ").strip()
                mgr.save(p); print("Đã lưu.")
            elif ch == "9":
                p = input("File: ").strip()
                mgr.load(p); print("Đã tải.")
            elif ch == "0":
                print("Bye!")
                break
            else:
                print("Chọn không hợp lệ.")
        except Exception as e:
            print("Lỗi:", e)

if __name__ == "__main__":
    main()
```
