# Tuần 3 : Matplotlib vẽ 10 biểu đồ

```python
import matplotlib.pyplot as plt
import numpy as np

# Dữ liệu mẫu
x = np.linspace(0, 10, 100)
y = np.sin(x)

# 1. Line plot
plt.figure()
plt.plot(x, y)
plt.title("Line Plot")
plt.show()

# 2. Bar chart
plt.figure()
plt.bar([1,2,3,4,5], [5,7,3,8,4])
plt.title("Bar Chart")
plt.show()

# 3. Scatter plot
plt.figure()
plt.scatter(np.random.rand(50), np.random.rand(50))
plt.title("Scatter Plot")
plt.show()

# 4. Histogram
plt.figure()
plt.hist(np.random.randn(1000), bins=30)
plt.title("Histogram")
plt.show()

# 5. Pie chart
plt.figure()
plt.pie([30, 20, 50], labels=["A", "B", "C"], autopct="%1.1f%%")
plt.title("Pie Chart")
plt.show()

# 6. Boxplot
plt.figure()
plt.boxplot(np.random.randn(100))
plt.title("Boxplot")
plt.show()

# 7. Stackplot
plt.figure()
days = np.arange(1,6)
plt.stackplot(days, [1,2,3,4,5], [1,4,6,8,10])
plt.title("Stackplot")
plt.show()

# 8. Stem plot
plt.figure()
plt.stem(range(10), np.random.rand(10))
plt.title("Stem Plot")
plt.show()

# 9. Filled area (area plot)
plt.figure()
plt.fill_between(x, y, color="skyblue", alpha=0.5)
plt.title("Area Plot")
plt.show()

# 10. Heatmap (dùng imshow)
plt.figure()
data = np.random.rand(10,10)
plt.imshow(data, cmap="viridis", interpolation="nearest")
plt.colorbar()
plt.title("Heatmap")
plt.show()
```
